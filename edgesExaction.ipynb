```python
import os
import numpy as np
import pandas as pd
import scipy.sparse as sp
import torch
data_dir='/data/NK/'
```

```python
data_adj=pd.read_csv(os.path.join(data_dir, f'ppi_of_NK.csv'),sep=",")
data_adj.head(10)
```

```python
index=data_adj.index
index
```

```python
ed=sp.coo_matrix(data_adj)
indices=np.vstack((ed.row,ed.col))
index=torch.LongTensor(indices)
values=torch.FloatTensor(ed.data)
edge_index=torch.sparse_coo_tensor(index,values,ed.shape)
edge_index=edge_index.coalesce().indices() 
edge_index
```

```python
def read_single_csv(input_path):
    df_chunk=pd.read_csv(input_path,sep=",",chunksize=5000) 
    res_chunk=[]
    for chunk in df_chunk:
        res_chunk.append(chunk)
    res_df=pd.concat(res_chunk)
    return res_df
```
